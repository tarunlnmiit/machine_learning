```python
API_KEY = ""
API_KEY_SECRET = ""
ACCESS_TOKEN = ""
ACCESS_TOKEN_SECRET = ""
```

```python
import tweepy
```

```python
auth = tweepy.OAuthHandler(API_KEY, API_KEY_SECRET)
```

```python
client = tweepy.Client(consumer_key=API_KEY, consumer_secret=API_KEY_SECRET, access_token=ACCESS_TOKEN, access_token_secret=ACCESS_TOKEN_SECRET)
```

```python
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
```

```python
api = tweepy.API(auth)
```

```python
text = 'This is a tweet from a bot. I am a bot!'
```

```python
response = client.create_tweet(text=text)
```

```python
response
```

```python
media = api.media_upload('test_image.png')
```
