```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/sales_records.csv')

print(df.shape)       # e.g., (15420, 12)
print(df.dtypes)
print(df.head())
print(df.describe())
print(df.isnull().sum())
print(df.columns)
```

```python
print(df.columns)

df.columns = ['_'.join(col.strip().lower().split(' ')).replace('(', '').replace(')', '') for col in df.columns.tolist()]
print(df.columns)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['rev_usd'] = pd.to_numeric(df['rev_usd'], errors='coerce')

df.head()
```

```python
revenue = df['rev_usd']
subset = df[['customer_id', 'rev_usd', 'region']]

print(revenue)
print(subset)
```

```python
high_value = df.loc[df['rev_usd'] > 10000, ['customer_id', 'rev_usd']]

high_value
```

```python
filtered = df[(df['region'] == 'North') & (df['rev_usd'] > 5000)]

filtered
```

```python
filtered_v2 = df.query("region == 'North' and rev_usd > 5000")
filtered_v2

```

```python
df[df['status'].isin(['active', 'trial'])]
```

```python
df['region'] = df['region'].str.strip().str.title()
df
```

```python
print(df['customer_id'].isnull().sum())
df_dropped = df.dropna(subset=['customer_id'])
df_dropped
```

```python
print(df.rev_usd.isnull().sum())
df_filled = df.fillna({'rev_usd': 0, 'notes': 'none'})
df_filled.rev_usd.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].median())

df_deduped = df.drop_duplicates(subset=['order_id'], keep='first')

df = df.rename(columns={
    'Customer ID ': 'customer_id',
    'Rev (USD)': 'revenue_usd'
})
```

```python
revenue_by_region = df.groupby('region')['rev_usd'].sum()
revenue_by_region
```

```python
region_summary = df.groupby('region').agg(
    total_revenue=('rev_usd', 'sum'),
    avg_order_value=('rev_usd', 'mean'),
    order_count=('order_id', 'count'),
    top_customer_revenue=('rev_usd', 'max')
).reset_index()
region_summary
```

```python
monthly_by_region = df.groupby(['region', 'month'])['rev_usd'].sum().reset_index()

monthly_by_region
```

```python
pivot = df.pivot_table(
    values='rev_usd',
    index='region',
    columns='month',
    aggfunc='sum',
    fill_value=0
)
pivot
```
