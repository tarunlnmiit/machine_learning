```python
# [start]:[end][:stride]


# Slice, Result
# 0:1, First item
# :1, First item(since default start is 0 & last index non-inclusive)
# :-2, From start until second to last item
# ::2, From start to end every second item
```

```python
mask = george > 7
mask
george[mask]
```

```python
# Operation, Example
# And, x[a & b]
# Or, x[a | b]
# Not, x[~a]
```

```python
mask2 = george <= 2
george[mask | mask2]
```
